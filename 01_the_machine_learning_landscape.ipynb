```julia
using Plots
using DataFrames
using Downloads
using CSV
using GLM;
```

```julia
# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = DataFrame(CSV.File(Downloads.download(data_root * "lifesat/lifesat.csv")))
X = lifesat[:, "GDP per capita (USD)"]
y = lifesat[:, "Life satisfaction"]

# Visualize the data
scatter(X, y, xlim=(23_500, 62_500), ylim=(4,9),
    xlabel="GDP per capita (USD)",
    ylabel="Life satisfaction"
)
```

```julia
# Select a linear model
fm = @formula(y ~ X)
model = lm(fm, DataFrame(X=X,y=y))
predict(model, DataFrame(X=37_655.2))
```
